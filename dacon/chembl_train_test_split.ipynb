```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# 학습 ChEMBL 데이터 로드
chembl_data = pd.read_csv('/content/drive/MyDrive/dacon/train.csv')
chembl_data.head()
```

```python
train, test = train_test_split(chembl_data, test_size=0.057)
```

```python
num_train = len(train['Smiles'].unique())
num_test = len(test['Smiles'].unique())

print('Number of Smiles')
print('Train:', num_train)
print('Test:', num_test)
```

```python
train.to_csv('/content/drive/MyDrive/dacon/preprocessed_train.csv', index=False)
```

```python
test.to_csv('/content/drive/MyDrive/dacon/preprocessed_test.csv', index=False)
```
